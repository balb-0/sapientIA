## Métodos de transformação das colunas

Os principais métodos de transformação de colunas são:

&nbsp;&nbsp;&nbsp;&nbsp; **Normalização**: esse método serve para ajustar os valores numéricos para que fiquem em uma escala padronizada, ocorre geralmente entre 0 e 1 e podendo variar entre -1 e 1, caso possua valores negativos, o que facilita a comparação entre diferentes variáveis com escalas discrepantes.

&nbsp;&nbsp;&nbsp;&nbsp; **Codificação**: transforma dados categóricos em números, o que permite o entendimento pelos modelos de machine learning.

&nbsp;&nbsp;&nbsp;&nbsp; **Padronização**: útil para ajustar os dados para que tenham média 0 e desvio padrão 1, bom quando os dados seguem uma distribuição normal e o objetivo é que todas as variáveis contribuam de forma semelhante.

&nbsp;&nbsp;&nbsp;&nbsp; **Transformação logarítmica**: tem objetivo de transformar variáveis que oscilam em escala exponencial para uma escala linear, geralmente utilizado em valores de renda.

### Histograma

&nbsp;&nbsp;&nbsp;&nbsp; Um histograma serve para visualizar a distribuição de uma variável numérica, mostrando como os dados estão distribuídos em diferentes intervalos. Ele permite identificar padrões como a dispersão, assimetria e presença de outliers, o que ajuda a entender melhor a forma da distribuição dos dados.

&nbsp;&nbsp;&nbsp;&nbsp; O histograma foi utilizado para verificar se os valores da coluna "Vl Liquido Final" seguem uma distribuição normal ou é assimétrico. Caso o gráfico siga o padrão de uma senoide, ele pode ser considerado como uma distribuição normal, ou próximo disso, mas nesse caso os valores estão majoritariamente na esquerda, o que indica ser assimétrico. Dessa forma, é possível inferir que a maior parte dos valores estão próximos de zero, o que reforça a não utilização do método de padronização.

### Boxplot


Um boxplot é usado para visualizar a distribuição de dados de uma variável e identificar possíveis outliers. Ele mostra a mediana, quartis (Q1 e Q3), e os valores mínimos e máximos, ajudando a entender a dispersão e a simetria dos dados, além de destacar valores extremos que podem influenciar a análise (os outliers).

Nesse caso, é possível notar que há vários outliers com valores altos. Além disso, a linha que indica a mediana (dentro do retângulo azul) está muito à esquerda em relação ao meio, o que indica a assimetria dos valores

## Transformação das colunas categóricas em colunas numéricas

Ao fazer um análise as colunas categóricas no DataFrame, foi decidido utilizar as seguintes colunas categóricas: **UEN, Segmento, Setor, Veículo**. Depois dessa etapa, foi necessário transformar esses dados para que pudessem ser processados pelo modelo de machine learning.

**One-Hot Encoding**

O One-hot encoding transforma cada categoria de uma coluna em colunas binárias separadas, onde cada coluna representa uma categoria com valores "0" ou "1". Este método é ideal quando o número de categorias é moderado ou baixo. A vantagem é que evita a introdução de uma ordem artificial, garantindo que o modelo não interprete de forma errada as categorias. No entanto, a desvantagem é o aumento significativo no número de colunas, especialmente se houver muitas categorias.

**Label Encoding**

O Label encoding converte cada categoria em um número inteiro único, atribuindo, por exemplo, "Rádio" como 1 e "Televisão" como 2. É útil quando as categorias possuem uma ordem implícita ou quando o espaço de memória é uma preocupação. A principal vantagem é a sua simplicidade e o fato de não aumentar o número de colunas. Contudo, ele pode introduzir uma ordem que não existe naturalmente, levando o modelo a possíveis interpretações errada.

**Ordinal Encoding**

O Ordinal encoding, similar ao label encoding, é utilizado quando as categorias têm uma ordem natural, como "Baixo", "Médio" e "Alto". Este método mantém a ordem hierárquica das categorias, o que pode ser benéfico para alguns modelos. A vantagem é a preservação da ordem, mas, como o label encoding, pode induzir um viés se a ordem não for interpretada corretamente pelo modelo.

**Frequency Encoding**

O Frequency encoding substitui cada categoria pela frequência com que aparece na coluna, transformando, por exemplo, "Rádio" em 0,2 se aparecer em 20% das observações. Esse método é útil quando a frequência das categorias está correlacionada com a variável alvo. Sua vantagem é a compacidade, sem aumentar o número de colunas, mas a desvantagem é que pode não capturar bem a informação se a frequência das categorias não for relevante para o alvo.

**Target Encoding**

Target encoding substitui cada categoria pela média da variável alvo associada a ela, como transformar "Rádio" em 500 se essa for a receita média associada a essa categoria. Ele é útil quando há uma correlação entre as categorias e a variável a ser predita. A vantagem é que pode melhorar o desempenho do modelo, mas há um risco de overfitting (comportamento indesejável de machine learning que ocorre quando o modelo de machine learning fornece previsões precisas para dados de treinamento, mas não para novos dados), se o método não for bem regularizado.


### Por que o Label Encoding?

O Label Encoding foi escolhido como método para transformar as colunas devido à sua capacidade de converter categorias em valores numéricos de maneira simples e eficiente, sem precisar criar novas variáveis. Este método atribui um número inteiro único para cada categoria, o que é especialmente útil quando se tem um grande número de categorias, como é o caso dessas colunas.

Portanto, o uso de label encoding foi escolhido como a técnica mais adequada para transformar as colunas categóricas neste projeto.

````python
from sklearn.preprocessing import LabelEncoder

# Instanciando o LabelEncoder
label_encoder = LabelEncoder()

# Aplicando o LabelEncoder às colunas
df1['UEN'] = label_encoder.fit_transform(df1['UEN'])

# Criando o dicionário que mapeia os valores originais para os números gerados
uen_dict = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

df1['Segmento'] = label_encoder.fit_transform(df1['Segmento'])
segmento_dict = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

df1['Setor'] = label_encoder.fit_transform(df1['Setor'])
setor_dict = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

df1['Veiculo'] = label_encoder.fit_transform(df1['Veiculo'])
veiculo_dict = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))


# Agora você tem os dicionários para cada coluna
print("UEN:", uen_dict)
print("Segmento:", segmento_dict)
print("Setor:", setor_dict)
print("Veiculo:", veiculo_dict)

````
